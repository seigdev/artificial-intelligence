```python
## removing stop words.

from nltk.stem import SnowballStemmer
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
```

```python
import nltk
nltk.download("stopwords")
```

```python
## returning stopwords based on language.
stopwords.words("english")
```

```python
# nltk.sent_tokenize(paragraph)
sentences = nltk.sent_tokenize(paragraph)
```

```python
snowballStemmer = SnowballStemmer("english")
lemmatizer = WordNetLemmatizer()
```

```python
## apply stopwords
## then filter
## apply stemming

for i in range(len(sentences)):
    # change sentence to lowercase.
    sentences[i] = sentences[i].lower()
    # perform word tokenization first.
    words = nltk.word_tokenize(sentences[i]) 
    # perform stemming if a word in the list of words not present in stopwords using list comprehension
    new_words = [lemmatizer.lemmatize(word, pos="n") for word in words if word not in set(stopwords.words("english"))]
    sentences[i] = " ".join(new_words) # converting all the new words into sentences
```
